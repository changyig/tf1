```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

from sklearn.tree import DecisionTreeClassifier,export_graphviz
import graphviz

import numpy as np
knn=KNeighborsClassifier(n_neighbors=1)
dataSet = load_iris()
x_train,x_test,y_train,y_test=train_test_split(dataSet['data'],dataSet['target'],random_state=0)
knn.fit(x_train,y_train)
x_new=np.array([[5,2.9,1,0.2]])
prediction =knn.predict(x_new)
y_pred =knn.predict(x_test)
# print(dataSet['data'])
print(prediction)
print(dataSet['target_names'][prediction])
print(np.mean(y_pred==y_test))
print(y_pred.shape)
print(y_pred)

# tree=DecisionTreeClassifier(random_state=0)
# tree.fit(x_train,y_train)
# export_graphviz(tree,out_file="tree.dot",class_names=["malignant","bengin"],feature_names=cancer.feature_names,impurity=False,filled=True)
# for i in y_pred :
#     print(dataSet['target_names'][i])
# data = dataSet['data'] # 数据
# label = dataSet['target'] # 数据对应的标签
# feature = dataSet['feature_names'] # 特征的名称
# target = dataSet['target_names'] # 标签的名称
# print(data.shape)
# print(dataSet.target)
```
